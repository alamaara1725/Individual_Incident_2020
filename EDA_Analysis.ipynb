```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Individual_Incident_2020-Copy1.csv")
print(df)
```

```python
#Which states have the highest number of incidents
plt.figure(figsize=(12,6))
df['state'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 States by Number of Incidents")
plt.xlabel("State")
plt.ylabel("Incident Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Are minor or non-minor victims more common
age_victim = [
    df['minor_victim'].sum(),
    df['non_minor_victim'].sum(),
    df['unknown_age_victim'].sum()
]

labels = ['Minor', 'Non-Minor', 'Unknown']

plt.figure(figsize=(7,5))
plt.bar(labels, age_victim)
plt.title("Victim Age Group Distribution")
plt.ylabel("Total Count")
plt.show()
```

```python
#Most common offense type
offense_cols = [
    'violence_offense','theft_offense','drug_offense',
    'sex_offense','kidnapping_trafficking','other_offense'
]

df[offense_cols].sum().sort_values().plot(kind='barh', figsize=(8,5))
plt.title("Total Offense Type Distribution")
plt.xlabel("Total Count")
plt.show()
```

```python
#Gender distribution of victims
victim_gender = [
    df['male_victim'].sum(),
    df['female_victim'].sum(),
    df['unknown_sex_victim'].sum()
]

labels = ['Male', 'Female', 'Unknown']

plt.figure(figsize=(6,6))
plt.pie(victim_gender, labels=labels, autopct='%1.1f%%')
plt.title("Victim Gender Distribution")
plt.show()
```
