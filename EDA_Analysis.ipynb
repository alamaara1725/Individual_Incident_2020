```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Individual_Incident_2020-Copy1.csv")
print(df)
```

```python
#How are records distributed across categories(Shows frequency of each category)
plt.figure(figsize=(5,10))
sns.countplot(y=df.select_dtypes(include='object').columns[0], data=df)
plt.title("Category-wise Record Distribution")
plt.show()
```

```python
#percentage share of top categories
#Displays contribution in percentage
top_cat = df.select_dtypes(include='object').columns[0]

df[top_cat].value_counts().head(5).plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.ylabel("")
plt.title("Top 5 Category Proportion")
plt.show()
```

```python
#shape of numerical data
num_col = df.select_dtypes(include='number').columns[0]

plt.figure(figsize=(8,5))
sns.histplot(df[num_col], kde=True)
plt.title(f"Distribution of {num_col}")
plt.show()
```

```python
#How does data change over time?

df['date_SIF'] = pd.to_datetime(df['date_SIF'], errors='coerce')

df.groupby(df['date_SIF'].dt.month).size().plot(marker='o')
plt.xlabel("Month")
plt.ylabel("Incident Count")
plt.title("Monthly Incident Trend")
plt.show()
```
